# Part 1: The Ecosystem — How the Tools Connect

Before installing anything, it helps to understand *why* each tool exists and how they work together. This section gives you that overview.

## 1.1 The Full Pipeline at a Glance


## 1.2 What Each Tool Does

### Miniconda
Miniconda is a package and environment manager for Python. Think of it as a toolbox organizer: instead of installing every Python library directly onto your computer (which can cause version conflicts), Miniconda creates isolated *environments* — each one is a self-contained workspace with its own set of tools.

We use Miniconda because PsychoPy requires **Python 3.10 specifically**, and your computer may already have a different Python version for other purposes.

### psychopy_env
This is the name of the conda environment we create for this project. It contains Python 3.10, PsychoPy, and all the packages that experiments depend on. The *recipe* for this environment is stored in a file called `psychopy_env.yml`, which is version-controlled in GitHub so every team member uses identical tools.

### VS Code
VS Code (Visual Studio Code) is a free code editor made by Microsoft. It is where you will read, edit, and run Python scripts. Key features we use:
- **Explorer panel** (left sidebar): Browse experiment files
- **Interpreter selector** (bottom-right status bar): Tells VS Code to use `psychopy_env`
- **Integrated terminal**: Run Python scripts without leaving the editor
- **GitHub Copilot** (optional): AI assistant for writing/editing code

### PsychoPy
PsychoPy is the Python library that actually *runs* cognitive tasks. It handles stimulus presentation, timing, keyboard input, and data saving. All five cognitive tasks used in the Lifespan Lab (N-back, task switching, stop-signal, digit symbol substitution, effort-based decision-making) are written using PsychoPy.

PsychoPy's gold-standard Python version is **3.10** as of 2026. This is why `psychopy_env` uses Python 3.10.

### Pavlovia
Pavlovia is an online platform that hosts PsychoPy experiments as web applications. When a PsychoPy task is pushed to Pavlovia via Git, the platform uses **PsychoJS** (a JavaScript version of PsychoPy) to run it in any web browser — including on smartphones. This is what makes EMA data collection possible.

### PsychoJS
PsychoJS is the translation layer that converts Python/PsychoPy experiments into JavaScript so they run in a browser. You do not write PsychoJS directly — Pavlovia handles the conversion automatically when you upload a task.

### m-Path
m-Path is the Experience Sampling (EMA) platform used to prompt participants throughout their day. It sends scheduled notifications to participants' smartphones. When a participant taps a notification, m-Path opens a link to the relevant Pavlovia-hosted task. This makes m-Path the *entry point* for participants in EMA studies.

### Local Lab Setting vs. EMA Setting
| Setting | Where task runs | Who controls timing | Data collection |
|---------|----------------|--------------------|-----------------|
| **Lab (Level 1)** | Research PC, PsychoPy | Researcher | Single session |
| **EMA (Level 2)** | Participant's phone, browser | m-Path notifications | Multiple times/day |

## 2.2 Step 1 — Fix Windows Script Execution Policy

Windows blocks scripts by default. You must change this **once** before Miniconda or Conda will work correctly in terminals.

1. Press `Win`, type **PowerShell**, right-click it → **Run as administrator**.
2. Paste and press `Enter`:

   ```powershell
   Set-ExecutionPolicy -ExecutionPolicy RemoteSigned -Scope CurrentUser
   ```

3. Type `Y` and press `Enter` if asked to confirm.
4. Close PowerShell.

## 2.7 Understanding the Two Terminal Modes

A common source of confusion: there are two different "modes" your terminal can be in.

**Mode 1 — Normal terminal (the front door)**
```
(psychopy_env) PS C:\Users\YourName\> _
```
This is where you run Python scripts and conda commands.

**Mode 2 — Python REPL (inside Python)**
```
>>>
```
You land here if you type `python` without a filename. This is an *interactive* Python interpreter — you can type Python commands directly, but you **cannot** run script files from here.

**If you see `>>>`, type `exit()` and press `Enter` to return to the normal terminal.**

# Part 3: Verify Your Installation (All Levels)

Before running real experiments, confirm that PsychoPy is working correctly.

## 3.1 PsychoPy Sanity Check

The code cell below will:
1. Open a PsychoPy window
2. Display **"PsychoPy is Working!"**
3. Wait 2 seconds and close

If you see the window appear and close cleanly, your setup is complete.

> **Note:** If you run this from inside Jupyter, you may need to run the cell **twice** on the first attempt — this is normal and due to how PsychoPy initializes its display backend.
>
> **Do not add `core.quit()` here.** In Jupyter, `core.quit()` would shut down the entire kernel. We use `win.close()` instead.

````python
from psychopy import visual
import time

# Create a window with a black background
win = visual.Window([800, 600], color='black', monitor='testMonitor')

# Create and display a text stimulus
text = visual.TextStim(win, text='PsychoPy is Working!', color='white', height=0.1)
text.draw()
win.flip()

# Wait 2 seconds, then close cleanly
time.sleep(2)
win.close()

print('Sanity check complete!')
````

## 3.2 What to Do if the Sanity Check Fails

| Error message | Most likely cause | Fix |
|--------------|------------------|-----|
| `ModuleNotFoundError: psychopy` | Wrong environment selected | Check interpreter (bottom-right of VS Code) — must say `psychopy_env` |
| Window opens then crashes instantly | Display driver issue | Try running the cell a second time |
| `FileNotFoundError` | Wrong working directory | In VS Code, open the experiment folder via File → Open Folder |

# Part 4: Platform Overview (Level 2 and 3)

This section introduces the platforms needed for deploying tasks online (Level 2) and for understanding the code well enough to modify it (Level 3).

## 4.1 GitHub

GitHub is a cloud platform that stores and tracks changes to code. It is how we:
- Share `psychopy_env.yml` so every team member uses identical tool versions
- Push task files to Pavlovia for online hosting
- Collaborate and keep a history of all changes

**Recommended resources:**
- [Pro Git Book](https://git-scm.com/book/en/v2) — comprehensive free reference
- [GitHub Interactive Skills](https://learn.github.com/skills/) — hands-on exercises
- [10-minute intro video](https://www.youtube.com/watch?v=a9u2yZvsqHA)

**Key concept:** A *repository* (repo) is a folder tracked by Git. Every change you make can be saved as a *commit*, creating a timeline of your work that you can always roll back.

## 4.2 Pavlovia

Pavlovia ([pavlovia.org](https://pavlovia.org/)) hosts PsychoPy experiments as web applications. When your task is pushed to Pavlovia via Git, the platform uses **PsychoJS** to run it in any web browser — including on participants' smartphones.

**Data flow:**
```
Your .py script  →  push to Pavlovia  →  Pavlovia converts to JS  →  runs in phone browser
```

## 4.3 PsychoJS

PsychoJS is the JavaScript library that makes PsychoPy experiments run in browsers. You do not write PsychoJS directly — Pavlovia handles the conversion. Understanding that it exists helps explain why some PsychoPy Python features (e.g. certain visual effects) behave differently in the web version.

## 4.4 m-Path

m-Path ([m-path.io](https://m-path.io/)) is the EMA (Ecological Momentary Assessment) platform. It:
- Sends scheduled notifications to participants' smartphones (e.g. 3–6 times per day)
- Embeds a link to the Pavlovia-hosted task in each notification
- Participants tap the link → their browser opens → they complete the task

m-Path is the *starting point* for participants — it connects the sampling schedule to the actual cognitive tasks.

## 4.5 Lab Setting vs. EMA Setting — Full Comparison

| | Lab Setting (Level 1) | EMA Setting (Level 2) |
|--|----------------------|----------------------|
| **Where task runs** | Research PC, PsychoPy | Participant's phone, browser |
| **Who triggers task** | Researcher | m-Path notification |
| **Task format** | Python (.py) | JavaScript (via PsychoJS/Pavlovia) |
| **Data storage** | Local .csv on PC | Cloud (Pavlovia) |
| **Session count** | Single session | Multiple per day, many days |
| **Controlled environment?** | Yes | No (real-world conditions) |

The `nBackAlpha3-master` folder in this repository contains a working cognitive task (spatial N-back working memory test). Use it to verify your full setup.

# Part 6: EMA Deployment Pipeline (Level 2)

This section explains how to move from a working local task to a task that participants can access on their smartphones.

## 6.1 Overview of Steps

```
Step 1: Verify task runs locally (Part 5)
    ↓
Step 2: Create a Pavlovia account and link your Git repo
    ↓
Step 3: Push task files to Pavlovia via Git
    ↓
Step 4: Set experiment to "Running" on Pavlovia dashboard
    ↓
Step 5: Copy Pavlovia task URL
    ↓
Step 6: Paste URL into m-Path study configuration
    ↓
Step 7: Test end-to-end (you as a test participant)
```

## 6.2 Push to Pavlovia

1. Ensure Git is installed (see Section 2.1, item 5).
2. On Pavlovia, create a new experiment and note the GitLab remote URL it provides.
3. In VS Code terminal, navigate to your experiment folder and run:

   ```cmd
   git init
   git remote add pavlovia <your-pavlovia-git-url>
   git add .
   git commit -m "Initial upload"
   git push pavlovia main
   ```

4. On the Pavlovia website, go to your experiment → change status from **Piloting** to **Running**.

## 6.3 Configure m-Path

1. Log in to m-Path and open your study.
2. In the notification or questionnaire settings, find the field for an external URL.
3. Paste the Pavlovia experiment URL.
4. Configure the notification schedule (time of day, frequency, days).
5. Save and activate the study.

## 6.4 End-to-End Test

Before recruiting participants:
1. Add yourself as a test participant in m-Path.
2. Receive a notification on your phone.
3. Tap it — the Pavlovia task should open in your phone's browser.
4. Complete the task.
5. Check that data appeared in the Pavlovia data dashboard.

> **Important:** Do not collect real participant data until you have successfully completed this end-to-end test.

# Part 7: Editing a Task (Level 3)

## 7.1 Before You Change Anything — Make a Backup

**Never edit the original task files without a backup.** The code below creates a timestamped backup of your experiment folder.

> Run this in a terminal (not in Jupyter) with `python backup.py`, or copy it to a standalone `.py` file.

````python
import shutil
import os
from datetime import datetime

# Set this to the folder you want to back up
SOURCE_FOLDER = os.path.join(os.path.dirname(os.getcwd()), 'nBackAlpha3-master', 'nBackAlpha3-master')

timestamp = datetime.now().strftime('%Y%m%d_%H%M%S')
backup_folder = SOURCE_FOLDER + f'_backup_{timestamp}'

shutil.copytree(SOURCE_FOLDER, backup_folder)
print(f'Backup created: {backup_folder}')
````

## 7.2 What You Can Safely Edit

| What to change | File to edit | Variable / location |
|----------------|-------------|---------------------|
| Stimulus display duration | `nBackAlpha3.py` | `time.sleep()` value in trial routine |
| Number of trials | `nBackAlpha3.py` | `nReps` in the trial loop |
| Target positions in conditions | `nBackCond1.xlsx`, `nBackCond2.xlsx`, `nBackCond3.xlsx` | `loc1`, `loc2`, `corrAns` columns |
| Stimulus appearance | `nBackAlpha3.py` | `color`, `size`, `pos` arguments |

## 7.3 Recommended Workflow for Edits

1. Make a backup (Section 7.1)
2. Make one small change at a time
3. After each change, run the sanity check (Part 3) to confirm PsychoPy still loads
4. Then run the task briefly to confirm your change behaves as intended
5. If something breaks, compare with your backup to identify what went wrong
