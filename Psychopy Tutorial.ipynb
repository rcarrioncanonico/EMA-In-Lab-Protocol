# How to Code In VS Code for Psychopy

In the world of coding, there are things called environments. Think of an environment like a dedicated kitchen where we store specific ingredients for our recipes.

A PsychoPy "recipe" (your experiment code) only works if it is cooked in the specific kitchen it was made for. In our case, the *psychopy_env* enviroment is necessary for us to be able to run and replicate the code, as it contains all the proper tools and ingredients for the recipe (code).

To work efficently in this kitchen, we need to install the PsychoPy enviroment inside of VS Code. This is because VS Code offers better tools to handle code such as Github Copilot, which will make it easier for editing and creating code. While there are different ways to achieve this, this tutorial focuses on Miniconda—a powerful tool that lets you manage multiple "kitchens" for different use cases (like one for PsychoPy and another for EEG data analysis).

You can go to this website: https://www.anaconda.com/docs/getting-started/miniconda/main to install Mini-Anaconda. Also, you will want to download the enviroment for Psychopy which you can find here: https://github.com/rcarrioncanonico/Momentary-Assesment-Tools/blob/main/psychopy_env.yml

Once installation is finished, we are going to run the newly added "Anaconda Prompt" application to create our new PsychoPy enviroment. You can look up the application by pression on the Windows Key and typing "Anaconda Prompt". In the prompt screen, we can run this commmand to create and install the PsychoPy enviroment:

    conda env create -f psychopy_env.yml

### Note ###
If you cannot find the psychopy_env.yml through the command above this means that you are not in the right directory. You can move your file to the directory that pops in the Anaconda Prompt or change the directory with this command, for example to move it to the Downloads folder:

     cd ~\Downloads
    
As seen previously in the Introduction Tutorial if you are not able to run commands in at the Anaconda Prompt this is mainly because of Windows security policy. You will need to deactivate. In order to do so run Powershell in administrator mode an]d run this command:

    Set-ExecutionPolicy -ExecutionPolicy RemoteSigned -Scope CurrentUser
