# How to Code In VS Code for Psychopy

In coding, an environment is like a dedicated kitchen where you keep specific ingredients for a recipe.

A PsychoPy "recipe" (your experiment code) only works if it is run in the kitchen it was made for. In our case, the `psychopy_env` environment is required because it contains the right tools and ingredients for the code.

To work efficiently in this kitchen, we need to set up the PsychoPy environment for VS Code. VS Code provides helpful tools like GitHub Copilot, which makes editing and creating code easier. There are different ways to do this, but this tutorial uses Miniconda so you can manage multiple environments (for example, one for PsychoPy and another for EEG analysis).

Install Miniconda here: https://www.anaconda.com/docs/getting-started/miniconda/main

Download the PsychoPy environment file here: https://github.com/rcarrioncanonico/Momentary-Assesment-Tools/blob/main/psychopy_env.yml

After installing Miniconda, open the "Anaconda Prompt" application to create the PsychoPy environment. You can find it by pressing the Windows key and typing "Anaconda Prompt". Then run:

```
conda env create -f psychopy_env.yml
```

### Note

If `psychopy_env.yml` is not found, you are in the wrong directory. Either move the file to the directory shown in Anaconda Prompt, or change directories. For example, to switch to Downloads:

```
cd ~\Downloads
```

As mentioned in the Introduction Tutorial, if you cannot run commands in Anaconda Prompt, it is likely due to Windows security policy. Open PowerShell as Administrator and run:

```
Set-ExecutionPolicy -ExecutionPolicy RemoteSigned -Scope CurrentUser
```

# Towards VS Code

You should now have an environment named `psychopy_env`. Open `miniconda3` > `envs` and confirm a `psychopy_env` folder exists.

In VS Code, set the interpreter and terminal to that same environment:

1) Interpreter: click the Python interpreter selector in the bottom-right status bar (next to the AI icon) and choose `psychopy_env`.
2) Terminal: open Terminal > New Terminal. It should activate `psychopy_env`. If it does not, run:

```
conda activate psychopy_env
```
